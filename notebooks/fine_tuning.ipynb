```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

model_path = r"G:\Ongoing Project\Fine Tuning LLM\my_mistral7b_finetuned_merged"

model = AutoModelForCausalLM.from_pretrained(
    model_path,
    device_map=None,
    torch_dtype=torch.float32,
    low_cpu_mem_usage=True,
    trust_remote_code=True,
    local_files_only=True,
    # These two lines are the real fix:
    load_in_4bit=False,      # ← force disable 4-bit
    load_in_8bit=False,      # ← force disable 8-bit
)

# Explicitly put everything on CPU
model = model.to("cpu")
model.eval()

tokenizer = AutoTokenizer.from_pretrained(model_path, local_files_only=True)

print("Your fine-tuned Mistral-7B is now loaded on CPU – no more quantization error!")
```
